```python
# Imports necessary libraries
import pandas as pd # Used to load, clean, and manipulate tabular data.
import numpy as np  # Handles numeric operations efficiently.
import matplotlib.pyplot as plt # The basic plotting tool.
import seaborn as sns # The pretty plotting tool—for cleaner, more insightful visualizations.
from datetime import datetime # For working with dates and times
import warnings
warnings.filterwarnings('ignore') # Silences Python’s whining about future warnings.
```

```python
df = pd.read_csv('owid-covid-data.csv') # loading the CSV file into a Pandas DataFrame called df
df # dfnow holds your entire COVID dataset in memory like a giant Excel sheet
```

```python
# change the datatype of the date column
df.date = pd.to_datetime(df.date) # Convert the 'date' column to a datetime format
df.info()
```

```python
# remove columns with 90% or more missing values.
miss_thresh = 0.9
df = df.dropna(thresh=(1 - miss_thresh) * len(df), axis=1)
df.info()

# len(df) = number of rows.
# (1 - miss_thresh) = 0.1 → So we keep columns with at least 10% of their values present.
# thresh=... sets the minimum non-null count required to keep a column.
# axis=1 says, “We’re chopping columns, not rows.”

#len(df) - How many rows exist
#df.shape - (Rows, Columns) structure
```

```python
cols_to_keep = ['iso_code', 'continent', 'location', 'date',
                'total_cases', 'new_cases', 'total_deaths', 'new_deaths',
                'total_tests', 'total_cases_per_million', 'new_tests',
                'positive_rate', 'male_smokers', 'female_smokers',
                'hospital_beds_per_thousand', 'people_vaccinated',
                'population_density', 'life_expectancy']

df = df[cols_to_keep]
df.info()
```

Activity 3: Worldwide COVID-19 Overview Providing a global and temporal perspective on COVID-19, identifying patterns and correlations crucial for strategic decision-making.
1. Visualize the WHO Regions (Continent column in dataset) with total COVID-19 cases and total deaths by using bar plots. .

2. Explore the world wide monthly trend of  COVID-19 cases using a line plot. .

3. Investigate the correlation between total cases and total deaths using a heatmap.

4. Analyze how total cases have evolved over time for a specific location (e.g., India).

```python
# Filter out rows where the continent is 'Unknown'
filtered_df = df[df['continent'] != 'Unknown']

# Group the filtered data by continent and calculate total cases and deaths
grouped_df = filtered_df.groupby('continent')[['total_cases', 'total_deaths']].sum()

# Display the grouped data
grouped_df

# filtered_df = df[df['continent'] != 'Unknown'] <> Removes any rows where the 'continent' value is 'Unknown'. This ensures only valid continent data is used.
# grouped_df = ...groupby('continent') <> Groups the filtered dataset by each unique continent.
# [['total_cases', 'total_deaths']].sum() <> Aggregates (sums up) the total number of cases and deaths for each continent.
# grouped_df <> Displays the final grouped and summed data.
```

```python
import matplotlib.pyplot as plt
grouped_df.plot(kind='bar')
plt.title('Total Cases and Deaths by Continent')
plt.xlabel('Continent')
plt.ylabel('Number of cases/deaths')
plt.show()

# grouped_df.plot(kind='bar') <> Creates a bar chart from the grouped_df DataFrame, where each bar represents total cases and deaths per continent.
# plt.title(...) <> Sets the title of the chart.
# plt.xlabel(...)<> Labels the X-axis as "Continent"
# plt.ylabel(...)<> Labels the Y-axis as "Number of cases/deaths"
# 	plt.show()<> Displays the plot window/output in Jupyter or Colab
```

```python
sns.lineplot(x='date', y='new_cases', data=df)
plt.title('monthly trend of new COVID-19 cases')
plt.xlabel('Date')
plt.ylabel('New of cases')
plt.xticks(rotation=45)
plt.show()

# sns.lineplot(...): Uses Seaborn to create a line chart showing how new COVID-19 cases change over time.
# x='date': Time is on the x-axis (horizontal).
# y='new_cases': Number of new cases is on the y-axis (vertical).
# data=df: Uses your full dataset (df) as the source of data.
# plt.title(...): Adds a chart title to explain what this line is showing.
# plt.xlabel(...): Labels the x-axis as “Date”.
# plt.ylabel(...): Labels the y-axis as “New of cases” (typo alert: it should say “Number of cases”).
# plt.show(): Actually displays the plot.
```

```python
corr_data = df[['total_cases', 'total_deaths']]

corr_matrix = corr_data.corr()

# Heatmap
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()

# df[['total_cases', 'total_deaths']] <> Extracts only the relevant columns for correlation
# corr() <> Calculates the Pearson correlation matrix between the selected columns
# sns.heatmap(...) <> Visualizes the correlation matrix with values (annot=True) and a blue-to-red color palette (cmap='coolwarm')
# plt.title() and plt.show() <> Add a title and render the plot.
```

```python
india_data = df[df['location'] == 'India']

# trend of total cases
sns.lineplot(x='date', y='total_cases', data=india_data)
plt.title('Trend of Total Cases in India')
plt.xlabel('Date')
plt.ylabel('Total Cases')
plt.xticks(rotation=45)
plt.show()
```
