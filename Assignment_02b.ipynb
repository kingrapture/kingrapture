## Data Quality: Missing Values and Data Types

Based on the `.isnull().sum()` and `.dtypes` outputs:

*   **Missing Values**:
    *   The 'Age' column has a significant number of missing values (177 out of 891).
    *   The 'Cabin' column has a large number of missing values (687 out of 891).
    *   The 'Embarked' column has a small number of missing values (2 out of 891).
    *   Other columns have no missing values.
*   **Data Types**:
    *   Numerical columns like 'PassengerId', 'Survived', 'Pclass', 'SibSp', and 'Parch' are integers (`int64`).
    *   'Age' and 'Fare' are floating-point numbers (`float64`), which is appropriate for continuous numerical data.
    *   'Name', 'Sex', 'Ticket', 'Cabin', and 'Embarked' are objects (`object`), indicating they are likely stored as strings or mixed types. This is expected for categorical data and free-form text.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create box plots for numerical features to identify outliers
numerical_cols = ['Age', 'Fare', 'SibSp', 'Parch', 'FamilySize']
plt.figure(figsize=(15, 10))
for i, col in enumerate(numerical_cols):
    plt.subplot(2, 3, i + 1)
    sns.boxplot(y=df[col])
    plt.title(f'Box Plot of {col}')
plt.tight_layout()
plt.show()
```

```python
# Chart numerical features appropriately
numerical_cols = ['Age', 'Fare', 'FamilySize']
plt.figure(figsize=(15, 5))
for i, col in enumerate(numerical_cols):
    plt.subplot(1, 3, i + 1)
    sns.histplot(df[col], kde=True)
    plt.title(f'Distribution of {col}')
plt.tight_layout()
plt.show()
```

```python
# Analyze the relationship between numerical features and the target variable ('Survived')
numerical_cols = ['Age', 'Fare', 'FamilySize']
plt.figure(figsize=(15, 5))
for i, col in enumerate(numerical_cols):
    plt.subplot(1, 3, i + 1)
    sns.boxplot(x='Survived', y=col, data=df)
    plt.title(f'{col} vs. Survived')
plt.tight_layout()
plt.show()
```
